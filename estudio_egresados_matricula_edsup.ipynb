```python
from utils import pipeline as pp
import pandas as pd
import glob
cnn=pp.get_conexion()
schema='egresados'
import numpy as np

```

```python
df_benchmark=pd.read_sql_query('select agno, rbd, cod_ense, mrun from egresados.fact_egresados where egresado_flg=1',cnn)
df_benchmark
```

# **Pruebas de Admisión**

```python
df_prueba=pd.read_csv('matricula_edsup/20230802_Matrícula_Ed_Superior_2012_PUBL_MRUN.csv',sep=';',encoding='utf8')
df_prueba=df_prueba[df_prueba['cat_periodo']==df_prueba['anio_ing_carr_ori']]
cols=['cat_periodo',  'mrun',
        'anio_ing_carr_ori', 
       'anio_ing_carr_act',  
       'tipo_inst_1'
       ,'tipo_inst_2',
       'tipo_inst_3'
       , 'cod_inst', 'nomb_inst', 
       'cod_carrera', 'nomb_carrera', 'modalidad', 'jornada',
       'dur_total_carr', 
         'nivel_carrera_1', 'nivel_carrera_2',
        'area_conocimiento',
       'cine_f_97_area', 'cine_f_97_subarea', 'area_carrera_generica',
       'cine_f_13_area', 'cine_f_13_subarea']
df_prueba=df_prueba.loc[:,cols]
df_prueba.dropna(subset=['mrun'],inplace=True)
df_prueba['mrun']=df_prueba['mrun'].astype('int64')
df_prueba.to_csv('bases/matricula_edsup/matricula_edsup_2012.csv')
df_prueba.info()
```

```python
lst_dfs=[]
for file in glob.glob('bases/matricula_edsup/*.csv'):
    lst_dfs.append(pd.read_csv(file,sep=','))
df_prueba=pd.concat(lst_dfs)
df_prueba.to_csv('matricula_edsup_2011_2023.csv',sep=';')
```

```python
df_matricula_edsup=pd.merge(df_prueba,df_benchmark,how='inner',on=['mrun']).iloc[:,1:]
df_matricula_edsup
                            
```

```python
tabla='matricula_edsup'
pp.insert_table(cnn,schema, tabla,df_matricula_edsup)
```
